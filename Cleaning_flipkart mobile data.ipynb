```python
import pandas as pd
```

```python
# Loading the dataset

df = pd.read_csv(r"D:/Data Analytics/PORTFOLIO_PROJECTS/_python_/web_scraping/flipkart_webscraping2.csv")
df
```

```python
# Deleting the Unnamed column
df = df.drop(columns=['Unnamed: 0'])
df
```

```python
# Removing special characters from Prices
df['Prices'] = df['Prices'].str.replace('₹', '').str.replace(',', '').fillna(0).astype(int)
df
```

```python
# Normalizing Description field by extracting "RAM","ROM" components
df['Description'] = df['Description'].str.split('|')
df['RAM'] = df['Description'].apply(lambda x: x[0].strip() if len(x) > 0 else '')
df['ROM'] = df['Description'].apply(lambda x: x[1].strip().split('ROM')[0] + 'ROM' if len(x) > 1 else '')

df
```

```python
# Deleting the original Description column
df = df.drop(columns=['Description'])
df
```

```python
# Handling Missing Values by filling it with "Unknown" or 0
df = df.fillna({'Product Name': 'Unknown', 'Prices': 0, 'Ratings': 0, 'RAM': 'Unknown', 'ROM': 'Unknown'})
df
```

```python
# List of known brands
known_brands = ["SAMSUNG", "OnePlus", "VIVO", "Xiaomi", "Realme", "OPPO", "Apple", "Nokia", "Motorola", "Poco", "Infinix", "Tecno", "Google", "Redmi"]

# Function to extract brand from "Product Name"
def extract_brand(product_name):
    for brand in known_brands:
        if product_name.upper().startswith(brand.upper()):
            return brand
    return 'Unknown'

# Applying the function to the 'Product Name' column to create the 'Brand' column
df['Brand'] = df['Product Name'].apply(extract_brand)
df
```

```python
# Removing Duplicates
df = df.drop_duplicates()
df
```

```python
# Rearranging the columns
df = df[['Product Name', 'Brand', 'Prices', 'RAM', 'ROM', 'Ratings']]
```

```python
# Saving the cleaned DataFrame to a new CSV file
df.to_csv("D:/Data Analytics/PORTFOLIO_PROJECTS/_python_/web_scraping/flipkart_cleaned.csv", index=False)
```
