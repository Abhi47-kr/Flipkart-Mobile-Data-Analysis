```python
# Importing necessary libraries

from bs4 import BeautifulSoup
import requests
import pandas as pd
```

```python
# Lists to store the extracted data

Product_name = []
Price = []
Description = []
Ratings = []
```

```python
# Loop through the first 10 pages of the search results
for i in range(1,278):
    
    # URL for the specific search query, with page number dynamically changing
    URL = f"https://www.flipkart.com/search?q=smart+phones&otracker=search&otracker1=search&marketplace=FLIPKART&as-show=on&as=off&page={i}"

    # Headers to mimic a real browser request 
    HEADERS = {"User-Agent": "Mozilla/5.0 (Windows NT 10.0; Win64; x64) AppleWebKit/537.36 (KHTML, like Gecko) Chrome/125.0.0.0 Safari/537.36 Edg/125.0.0.0", "Accept-Encoding":"gzip, deflate", "Accept":"text/html,application/xhtml+xml,application/xml;q=0.9,*/*;q=0.8", "DNT":"1","Connection":"close", "Upgrade-Insecure-Requests":"1"}

    # Requesting the content of the page
    page = requests.get(URL, headers=HEADERS)

    # Parsing the content of the page with BeautifulSoup
    soup = BeautifulSoup(page.content, "html.parser")

    # Finding the main container that holds the product listings
    box = soup.find("div",class_="DOjaWF gdgoEp")

    # Extracting each product's container within the main container
    product_container = box.find_all("div", class_="tUxRFH")

    # Loop through each product container to extract details
    for container in product_container:

        # Extract and append the product name
        name = container.find("div",class_="KzDlHZ")
        Product_name.append(name.text.strip() if name else "")
        
        # Extract and append the product price
        pricing = container.find("div", class_="Nx9bqj _4b5DiR")
        Price.append(pricing.text.strip() if pricing else "")
        
        # Extract and append the product ratings    
        descript = container.find("ul", class_="G4BRas")
        Description.append(descript.text.strip() if descript else "")
        
        # Extract and append the product ratings
        rat = container.find("div", class_="XQDdHH")
        Ratings.append(rat.text.strip() if rat else "")
        
```

```python
# Creating a DataFrame from the extracted data

df = pd.DataFrame({
    "Product Name":Product_name,
    "Prices":Price,
    "Description":Description,
    "Ratings":Ratings
})

# Display the DataFrame (for verification)
df
```
